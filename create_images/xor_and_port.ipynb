```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
X = [[0,0], [0,1], [1,0], [1,1]]
Y_xor = [0, 1, 1, 0]
Y_and = [0, 0, 0, 1]
c = ['b', 'r']
```

```python
plt.figure(dpi=500, figsize=(3,3))

for x,y in zip(X, Y_xor):
    plt.plot(x[0], x[1], '.', color=c[y])
plt.plot([-0.5, 1], [1, -0.5], '--g')
plt.plot([0, 1.5], [1.5, 0], '--g')

plt.xlabel("X1")
plt.ylabel("X2")
axes = plt.gca()
axes.fill_between([-0.5, 0, 1, 1.5], [1, 0.5, -0.5, -1], [2, 1.5, .5, 0], facecolor='c', alpha=0.4)
axes.set_ylim([-.5,1.5])
axes.set_xlim([-.5,1.5])
plt.show()

```

```python
plt.figure(dpi=500, figsize=(3,3))

for x,y in zip(X, Y_and):
    plt.plot(x[0], x[1], '.', color=c[y])
plt.plot([0, 1.5], [1.5, 0], '--c')

plt.xlabel("X1")
plt.ylabel("X2")
axes = plt.gca()
axes.set_ylim([-.5,1.5])
axes.set_xlim([-.5,1.5])
plt.show()
```

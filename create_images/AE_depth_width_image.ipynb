```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
```

```python
fig = plt.figure(figsize = (5,3),dpi=250)
ax = fig.gca(projection='3d')
x = range(1,6)
y = [2126, 1063, 532, 266, 133]
X, Y = np.meshgrid(x, y)
Z = [[163.6, 41.7, 30.4, 37.2, 80.9],
     [122.9, 36.1, 29.0, 37.7, 64.9],
     [ 66.6, 34.1, 29.0, 32.2, 48.9],
     [ 67.1, 44.2, 32.4, 40.8, 45.2],
     [ 60.3, 45.2, 39.1, 41.9, 51.5]]
Z = np.array(Z)

surf = ax.plot_surface(X, Y, Z, cmap=cm.rainbow,
                       linewidth=10, antialiased=True,
                       alpha = 1)

ax.set_xlim(1, 5)
ax.set_ylim(0, 2126)
ax.set_zlim(20, 140)
# ax.set_xticklabels(x, rotation=0, fontsize=8)
# ax.set_yticklabels(y, rotation=0, fontsize=8)
ax.set_xlabel(r"depth", fontsize=10)
ax.set_ylabel(r"height", fontsize=10)
ax.set_zlabel(r"$\Delta$ (-)", fontsize=10)
ax.view_init(50, -35)
ax.dist = 13
plt.rcParams['xtick.labelsize']=9.4

fig.colorbar(surf, aspect=20, shrink=0.7)
plt.show()
```

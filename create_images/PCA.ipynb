```python
import numpy as np
import matplotlib.pyplot as plt
import random
import scipy.linalg as linalg
```

```python
x = [random.random()*3 for _ in range(50)]
y = [2*i + (random.gauss(0, 0.5)) for i in x]
```

```python
plt.figure(dpi=500, figsize=(3,3))
plt.plot(x,y,'.')
plt.plot([0,3], [0,6], '--r')
plt.xlabel("x")
plt.ylabel("y")
axes = plt.gca()
axes.set_xlim([0,3])
plt.show()
```

```python
data = np.array([x,y]).T
data = data - np.mean(data, axis = 0)
plt.plot(data.T[0], data.T[1], '.')
```

```python
TransformationMatrix = np.dot(data.T, data)
sx = np.std(x)
sy = np.std(y)

# TransformationMatrix[0][0] = TransformationMatrix[0][0]/(sx**2)
# TransformationMatrix[1][1] = TransformationMatrix[1][1]/(sy**2)
# TransformationMatrix[0][1] = TransformationMatrix[0][1]/(sx*sy)
# TransformationMatrix[1][0] = TransformationMatrix[1][0]/(sx*sy)

print(TransformationMatrix / 50)
eig = np.real(linalg.eig(TransformationMatrix)[0])
TransformationMatrix = np.real(linalg.eig(TransformationMatrix)[1])
print(TransformationMatrix)
```

```python
plt.figure(dpi=500, figsize=(3,3))
plt.plot(*np.dot(data, TransformationMatrix[:,::-1]).T, '.')
plt.xlabel("principle component 1")
plt.ylabel('principle component 2')
axes = plt.gca()
axes.set_ylim([-4.5,4.5])
axes.set_xlim([-4.5,4.5])
plt.show()
```

```python
plt.figure(dpi=500, figsize=(6,1))
plt.plot(np.dot(data, TransformationMatrix[:,1]), [0] * 50, '.')
plt.xlabel("principle component 1")
axes = plt.gca()
axes.set_ylim([-1,1])
axes.set_xlim([-4.5,4.5])
plt.show()
```

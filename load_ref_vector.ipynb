```python
wavelength = np.load("../data/Raman/wavelength.npy", 'r')

files = ['blue0', 'blue1', 'blue2', 'blue3', 'blue4',
         'green0', 'green1', 'green2',
         'red0', 'red1', 'red2']

ref_vectors = {}
for file in files:
    ref_vectors[file] = np.load(f"../data/Raman/{file}.npy", 'r')
```

```python
plt.figure(figsize = (20,12))
for key, value in ref_vectors.items():
    if 'blue' in key:
        plt.plot(wavelength, value, label=key)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (20,12))
for key, value in ref_vectors.items():
    if 'red' in key:
        plt.plot(wavelength, value, label=key)
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# checking algorithm from the .py
file = 'Alina_art_1_1_RS.txt'

df = pd.read_csv('signal_split_manios_psillakis/' + file, delimiter='\t', skipinitialspace=True, header=None, skiprows=[])
wavelength = df.iloc[0].to_numpy()[2:]
data = df.iloc[1:].to_numpy()

X = list(np.unique(data[:,0]))
Y = list(np.unique(data[:,1]))

img = np.empty((len(X), len(Y), len(wavelength)), dtype=np.float64)

for d in data:
    i = X.index(d[0])
    j = Y.index(d[1])
    img[i,j,:] = d[2:]
```

```python
df = pd.read_csv('../../data/Raman/raw/' + file, delimiter='\t', skipinitialspace=True, header=None, skiprows=[])
wavelength = df.iloc[0].to_numpy()[2:]
data = df.iloc[1:].to_numpy()

print(data.shape)


X = list(np.unique(data[:,0]))
Y = list(np.unique(data[:,1]))

img = np.empty((len(X), len(Y), len(wavelength)), dtype=np.float64)

for d in data:
    i = X.index(d[0])
    j = Y.index(d[1])
    img[i,j,:] = d[2:]
```

```python
X = img.reshape(-1, len(wavelength))
X.shape
```

```python
df2 = pd.DataFrame(index=df['x'].unique(), columns = df['y'].unique())
img = np.empty((df['x'].nunique(), df['y'].nunique(), df['wavelength'].nunique()))

for i, x in enumerate(df['x'].unique()):
    for j, y in enumerate(df['y'].unique()):
        img[i,j,:] = df[(df['x']==x) & (df['y']==y)]['value'].to_numpy()
```

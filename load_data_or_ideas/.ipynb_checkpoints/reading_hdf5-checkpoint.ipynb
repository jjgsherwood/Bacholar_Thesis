```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = ["../../data/Kidney_map/Kidney_map_150z25_60s_#37b_f_(0_0)_Ref.TippRHSIRef",
            "../../data/Kidney_map/Kidney_map_150z25_60s_#37b_f_PL_(0_0)_Ref.TippRHSIRef",
            "../../data/Kidney_map/Kidney_map_150z25_60s_#37b_f_RS_(0_0)_Ref.TippRHSIRef"]

data = []
for file in filename:
    with h5py.File(file, "r") as f:
        # List all groups
        # print("Keys: %s" % f.keys())
        a_group_key = list(f.keys())[0]

        # Get the data
        data.append(np.array(f[a_group_key]))
```

```python
plt.figure(figsize=(14,10))
for d in data:
    plt.plot(d)
```

```python
np.save(f'../../data/Kidney_map/Kidney_map_150z25_60s_#37b_f_(0_0)_Ref.npy', data[0]) 
```
